```python
#builtin pandas data visualization
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})

df1 = pd.read_csv("df1", index_col=0)
df2 = pd.read_csv("df2")
```

```python
df1.head()
```

```python
df1['A'].hist(bins=30) #plotting a hist
```

```python
df1['A'].plot(kind="hist", bins=30) #plotting a hist alternate way
```

```python
df1['A'].plot.hist(bins=30)#plotting a hist in another way
```

```python
df2.head()
```

```python
df2.plot.area(alpha=0.4) #plotting an area of points
```

```python
df2.plot.bar(stacked=True) #plotting a bar
```

```python
df1.plot.line(y='B',figsize=(12,3),lw=1) #plotting a line graph
```

```python
df1.plot.scatter(x='A', y='B', c='C', cmap="coolwarm") #scatter
```

```python
df1.plot.scatter(x='A', y='B', s=df1['C']*30)
```

```python
df2.plot.box() #boxplot
```

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
```

```python
df.plot.hexbin(x='a', y='b', gridsize=25) #hexplot
```

```python
df2.plot.kde() #kde plot
```
