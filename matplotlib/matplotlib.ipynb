```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
#creating a dataset
```

```python
x
```

```python
y
```

```python
#Functional method
plt.plot(x, y, "r*")
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.title("A Title")
```

```python
#multiplots on the same canvas
plt.subplot(1, 2, 1)
plt.plot(x, y, "r+")
plt.subplot(1,2,2)
plt.plot(y, x, "b+")
```

```python
#Object Oriented method
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel("X-label")
axes.set_ylabel("Y-label")
axes.set_title("A Title")
```

```python
#axes inside an axes
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_xlabel("X-label")
axes1.set_ylabel("Y-label")
axes1.set_title("A Title")

axes2.plot(y, x)
axes2.set_xlabel("X-label")
axes2.set_ylabel("Y-label")
axes2.set_title("Another Title")
```

```python
#subplots using object oriented method
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[0].set_xlabel("X-label")
axes[0].set_ylabel("Y-label")
axes[0].set_title("A Title")

axes[1].plot(y, x)
axes[1].set_xlabel("X-label")
axes[1].set_ylabel("Y-label")
axes[1].set_title("Another Title")
plt.tight_layout()
```

```python
#figure size, aspect ratio and dpi
fig = plt.figure(figsize=(8, 2), dpi=100)
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8, 2), dpi=500)

ax[0].plot(x, y)
ax[1].plot(y, x)
plt.tight_layout()
```

```python
#saving a figure
fig.savefig("saved_figure.png", dpi=500)
```

```python
#two plots in one
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.plot(y, x)
ax.set_xlabel("X-label")
ax.set_ylabel("Y-label")
ax.set_title("A Title")
```

```python
#making legends
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label="xSquared")
ax.plot(x, x**3, label="xCubed")
ax.legend(loc="best")
```

```python
#plot  style and apperance
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color="green", lw=20, ls=":")
ax.plot(x, x, color="blue", linestyle="dotted")
ax.plot(y, x, color="#FF8C00", linewidth=3, alpha=0.4, marker="x", markersize=20)
ax.set_xlabel("X-axes")
ax.set_ylabel("Y-axes")
ax.set_title("Styled Plot")
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
```

```python
#scatter plots
plt.scatter(x, y)
```

```python
#histogram plot
from random import sample
data = sample(range(1, 1000), 100)
plt.hist(data)
```

```python
#boxplot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, vert=True, patch_artist=True)
```

```python
#sample data
import numpy as np
x = np.arange(0, 100)
y = x * 2
z = x ** 2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
#importing and setting inline
```

```python
#plotting using oop method
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_title("title")
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
#plotting using oop method with two axes
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
ax1.plot(x, y, color="blue")
ax2.plot(y, x, color="pink")
```

```python
#plotting using oop method with two axes and x lims
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
ax1.plot(x, x**2, color="blue")
ax2.plot(x, y, color="pink")
ax2.set_xlim([20, 22])
ax2.set_ylim([30, 50])
```

```python
#creating subplots
fig, ax = plt.subplots(figsize=(12, 2), nrows=1, ncols=2)
ax[0].plot(x, y, color="blue", ls='-')
ax[1].plot(x, x**2, color="red", lw=5)
```
