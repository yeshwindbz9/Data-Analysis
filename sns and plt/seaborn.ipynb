#Seaborn is a stastical plotting library, it is designed to work well with the pandas dataframe objects. checkout github.com/mwaskom/seaborn to know more

```python
import pandas as pd
import seaborn as sns #importing seaborn
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})
import matplotlib.pyplot as plt
plt.rcParams["patch.force_edgecolor"] = True #for bins border
```

```python
#loading a builtin dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python
#distribution plots, is univariant, allows to visualize the distribution
sns.distplot(tips["total_bill"], kde=True, bins=20)
#notice we get a kde, a kernel density estimation and a histogram
#by analyzing we know that most bills are from 10 to 20 $
```

```python
#joint plot, is bivariant and allows us to compare two distribution plots
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
```

```python
#pairplot, does jointplot for every possible numeric feature in the dataframe
sns.pairplot(tips, hue="sex") #use hue for categorical feature
```

```python
#rugplot, shows the frequency of points in the specified feature
sns.rugplot(tips["total_bill"])
```

```python
#Categorical plots, see distributions for categorical features wrt numerical features
tips.head()
```

```python
import numpy as np
import seaborn as sns
```

```python
#barplot, aggregrates categorical data based on the estimator, by default it's mean
sns.barplot(x="sex", y="total_bill", data=tips, estimator=np.mean)
```

```python
#countplot, same as barplot but estimator counts the number of instances
sns.countplot(x="sex", data=tips)
```

```python
#boxplot/whiskerplot, shows distribution between variables
sns.boxplot(x="day", y="total_bill", data=tips, hue="smoker")
#the boxes shows the quartile, while the whiskers extends to show rest of the distribution called outliers
```

```python
#violin plot, shows distribution of data based on a category
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split=True)
```

```python
#strip plot, just like scatter plot
sns.stripplot(x="day", y="total_bill", hue="sex", split=True, data=tips, jitter=True)
```

```python
#swarmplot, combines strip and violin plot
sns.swarmplot(x="day", y="total_bill", hue="sex", split=True, data=tips)
```

```python
#factor plot, general form of all plots, use kind to represent different plots
sns.factorplot(x="day", y="total_bill", hue="sex", kind="bar", data=tips)
```

```python
#factor plot, general form of all plots, use kind to represent different plots
sns.factorplot(x="day", y="total_bill", hue="sex", kind="violin", data=tips)
```

```python
#factor plot, general form of all plots, use kind to represent different plots
sns.factorplot(x="day", y="total_bill", hue="sex", data=tips)
```

```python
#matrix plots & heat maps
flights = sns.load_dataset("flights")
flights.head()
```

```python
tc = tips.corr() #getting data in matrix format
```

```python
sns.heatmap(tc, annot=True, cmap="coolwarm") #calling heatmap on it
#colors those values based on dataframe
```

```python
fp = flights.pivot_table(index="month", columns="year", values="passengers")
fp
```

```python
sns.heatmap(fp, cmap="coolwarm", lw=3, linecolor="white") #will give density of the plots
```

```python
#clustermap, makes use of clustering to make a clustered heatmap
sns.clustermap(fp, cmap="coolwarm", standard_scale = 1)
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
#to quickly check the unique values
iris["species"].unique()
```

```python
#pairplot, is an automated jointplot for the dataset
sns.pairplot(iris)
#using pair grid we can customize this graph for a lot more control
```

```python
#PairGrid
g = sns.PairGrid(iris)
#g.map(plt.scatter)
g.map_diag(sns.displot)
g.map_upper(sns.displot)
g.map_lower(sns.kdeplot)
```
